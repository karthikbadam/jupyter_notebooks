```python
import wikia
import json
import sys
import requests

from httpstream import get

from bs4 import BeautifulSoup

soup = BeautifulSoup

root_url = 'http://marvel.wikia.com'

marvelArticles = requests.get('http://marvel.wikia.com/api/v1/Articles/Top')

itemsList = json.loads(marvelArticles.text).get('items')

#print (itemsList)

x = []
y = [] 

for item in itemsList:
    
    itemTitle = item.get('title')
    
    itemUrl = root_url + item.get('url')
    
    #print (itemTitle + ":" + itemUrl)
    
    # get links to these top pages
    itemHtml = requests.get(itemUrl)
    
    soupPage = BeautifulSoup(itemHtml.text)
    
    itemLinks = soupPage.find_all("a")
    
    count = 0
    
    for itemLink in itemLinks:
        
        count+=1    
    
    x.append(itemTitle)
    y.append(count)
       
    print itemTitle + " has {} links".format(count)
    
```
